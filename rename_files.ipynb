```python
import pandas as pd
import os
from fuzzywuzzy import fuzz, process
import re
```

```python
def rename_files_in_directory(directory_path, start_index=100):
    # List all files in the directory
    files = os.listdir(directory_path)
    
    # Sort files to maintain a consistent order
    files.sort()

    # Iterate over the files and rename them with the specified start index
    for idx, filename in enumerate(files, start=start_index):
        # Get the file extension
        file_extension = os.path.splitext(filename)[1]
        
        # Create new filename
        new_filename = f"{idx}{file_extension}"
        
        # Construct full file paths
        old_file = os.path.join(directory_path, filename)
        new_file = os.path.join(directory_path, new_filename)
        
        # Rename the file
        os.rename(old_file, new_file)
        print(f"Renamed: {old_file} -> {new_file}")
```

```python
rename_files_in_directory('data/non_hs_extension', start_index=100)
```
